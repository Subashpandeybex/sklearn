# visulaization

```python
def plot(dataX, dataY, ix=0, iy=1):
    target_names = ['sepal length (cm)', 'sepal width (cm)', 'petal length (cm)', 'petal width (cm)']
    for i, target in enumerate(dataY):
        x, y = dataX[i][ix], dataX[i][iy]
        t = target
        scale = 100.0
        color = ""
        if t == 0:
            color = "yellow"
        elif t == 1:
            color = "red"
        elif t == 2:
            color = "green"

        plt.scatter(x, y, c=color, alpha=1, edgecolor="none")

    # Legend ( the call signiture corresponding)
    red_patch = mpatches.Patch(color='yellow', label='iris setosa')
    green_patch = mpatches.Patch(color='red', label='iris versicolor')
    blue_patch = mpatches.Patch(color='green', label='iris virginica')
    #plt.legend(handles=[red_patch, green_patch, blue_patch])

    plt.title("The Iris Data Set", fontsize=18)
    plt.xlabel(target_names[ix], fontsize=15)
    plt.ylabel(target_names[iy], fontsize=15)

    plt.legend(handles=[red_patch, green_patch, blue_patch])
    plt.grid(True)
    


```

```python
plot(x,y)
```

### choosing value of k 

```python
# this estimate the accuracy
from sklearn.model_selection import cross_val_score # evaluating the estimator performance

import numpy as np
```

```python
# creating odd list of K for KNN
klist = list(range(1,50, 2))

# empty list that will hold cv scores
cv_scores = []
```

```python
# perform 10-fold cross validation for each k value
for k in klist:
    knn = KNeighborsClassifier(n_neighbors=k)
    scores = cross_val_score(knn, x, y, cv=10, scoring='accuracy')
    cv_scores.append(scores.mean())
```

```python
# changing to misclassification error
errors = 1- np.array(cv_scores) # 1-accuracy
```

```python
# determining best k
optimal_k = klist[np.argmin(errors)]  # gives the minimum value
print("The optimal number of neighbors is {}".format(optimal_k))
```

```python
# plot misclassification error vs k
plt.plot(klist, errors)
plt.xlabel('Number of Neighbors K')
plt.ylabel('Misclassification Error')
plt.show()
```
